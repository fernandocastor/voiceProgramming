```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt

# The following Pandas option sets the maximum width of a column when presenting tabulated data to be 200.
pd.options.display.max_colwidth = 200
# The line above makes matplotlib inline the graphics is produces.
%matplotlib inline
```

```python
corpus = ['The sky is blue and loving and beautiful.',
          'Love this blue and beautiful sky!',
          'The quick brown fox jumps over the lazy dog.',
          "A king's breakfast has sausages, ham, bacon, eggs, toast and beans",
          'I love green eggs, ham, sausage and bacon!',
          'The brown fox is quick quick and the blue dog is lazy laziness!',
          'The sky is very blue and the sky is beautifully beautiful today',
          'The dog is lazy but skies the brown fox is quick dog!'    
]
labels = ['weather', 'weather', 'animals', 'food', 'food', 'animals', 'weather', 'animals']

corpus = np.array(corpus)
corpus_df = pd.DataFrame({'Document': corpus, 'Category': labels})
corpus_df = corpus_df[['Document', 'Category']]
corpus_df
```

```python
"""
Preprocessing text involes multiple activities:
- Removing tags (for HTML)
- Removing accented characters
- Expanding contractions ("don't" => "do not")
- Removing special characters
- Stemming and lemmatization
- Removing stopwords
"""
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')

def normalizeDocument(doc): 
    # flags re.I and re.A below refer to re.ASCII and re.IGNORECASE (https://docs.python.org/3/library/re.html)
    procDoc = re.sub(r'[^a-zA-Z\s]', '', doc, re.I|re.A)
    procDoc = procDoc.lower()
    procDoc = procDoc.strip()
    tokens = wpt.tokenize(procDoc)
    filteredTokens = [token for token in tokens if not(token in stop_words)]
    
    lemmatizer = nltk.wordnet.WordNetLemmatizer()
    filteredLemmatizedTokens = [lemmatizer.lemmatize(token) for token in filteredTokens]
    ps = nltk.porter.PorterStemmer()
    filteredLemmatizedStemmedTokens = [ps.stem(token) for token in filteredLemmatizedTokens]
    
    
    procDoc = ' '.join(filteredLemmatizedStemmedTokens)

    return procDoc

# Don't completely understand vectorize, but I know that using it I can process the array of strings 
# directly, without the need to take each element and process it separately. 
normalizeCorpus = np.vectorize(normalizeDocument)

nCorpus = normalizeCorpus(corpus)
nCorpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Extracting features from the text using the bag of words model. For each string in the corpus, 
# it createses a vector whose number of elements is the number of unique words in the corpus and
# each cell of said vector corresponds to one of these words and indicates the number of times
# that word appears in the string.
cv = CountVectorizer(min_df=0., max_df=1.)
cvMatrix = cv.fit_transform(nCorpus)
cvMatrix = cvMatrix.toarray()
cvMatrix
```

```python
# Presenting the feature array in a more organized manner.
vocab = cv.get_feature_names()
pd.DataFrame(cvMatrix, columns=vocab)
```

```python
# ngram_range=(X,Y) => X is the minimum and Y is the maximum number of elements in the N-gram.
bv = CountVectorizer(ngram_range=(2,2))
bvMatrix = bv.fit_transform(nCorpus)

bvMatrix = bvMatrix.toarray()
vocab = bv.get_feature_names()
pd.DataFrame(bvMatrix, columns=vocab)
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=True)
tvMatrix = tv.fit_transform(nCorpus)
tvMatrix = tvMatrix.toarray()

vocab = tv.get_feature_names()
pd.DataFrame(np.round(tvMatrix, 2), columns=vocab)
```

```python
# To gauge document similarity
from sklearn.metrics.pairwise import cosine_similarity

similarityMatrix = cosine_similarity(tvMatrix)
similarityDF = pd.DataFrame(similarityMatrix)
similarityDF
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

#  "Ward's minimum variance method" is employed as the linkage criterion to minimize total within-cluser variance
Z = linkage(similarityMatrix, 'ward')
pd.DataFrame(Z, columns=['Document\Cluster 1', 'Document\Cluster 2', 
                         'Distance', 'Cluster Size'], dtype='object')
```

```python
plt.figure(figsize=(8, 3))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
dendrogram(Z)
plt.axhline(y=1.0, c='k', ls='--', lw=0.5)
```

```python
from scipy.cluster.hierarchy import fcluster 
maxDist = 1.0

clusterLabels = fcluster(Z, maxDist, criterion='distance')
clusterLabels = pd.DataFrame(clusterLabels, columns=['ClusterLabel'])
pd.concat([corpus_df, clusterLabels], axis=1)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

#n_components is the number of topics
lda = LatentDirichletAllocation(n_components=3, max_iter=10000, random_state=0)
# dt => document-term
dtMatrix = lda.fit_transform(cvMatrix)
features = pd.DataFrame(dtMatrix, columns=['T1', 'T2', 'T3'])
features
```

```python
ttMatrix = lda.components_
for tWeights in ttMatrix: 
    topic = [(token, weight) for token, weight in zip(vocab, tWeights)]
    topic = sorted(topic, key=lambda x: -x[1])
    topic = [item for item in topic if item[1] > 0.6]
    print(topic)
    print()
```

```python
# Another clustering method besides the one we used previously, hierarchical clustering. 
# For this specific example, K-Means will create the same clusters as the other approach.
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=0)
km.fit_transform(features)
clusterLabels = km.labels_
clusterLabels = pd.DataFrame(clusterLabels, columns=['ClusterLabel'])
pd.concat([corpus_df, clusterLabels], axis=1)
```
